```python
import requests
import json
```

```python
# https://api.github.com/

#список репозиториев для пользователя и сохр как json

USER = 'alex-zotov'
url = f'https://api.github.com/users/{USER}/repos'

resp = requests.get(url)

print(resp.status_code)

repos = resp.json()
print(repos)
```

```python
# Сохраняем в JSON файл
with open(f'git_repos_{USER}.json', 'w') as f:
        json.dump(repos, f, indent=2)
```
