```python
from telethon import TelegramClient
from telethon.tl.types import User
import telegram_key
import asyncio
```

```python
# https://docs.telethon.dev/en/stable/basic/signing-in.html
# –ª–æ–≥–∏–Ω–∏–º—Å—è –Ω–∞ https://my.telegram.org/
# —Ä–µ–≥–∞–µ–º –ø—Ä–∏–ª–æ–∂–µ–Ω–∏–µ API development tools https://my.telegram.org/apps
# –ø–æ–ª—É—á–∞–µ–º API_ID int, API_HASH string
```

```python
# –ø–æ–ª—É—á–∏—Ç—å —Å–æ–æ–±—â–µ–Ω–∏—è —á–∞—Ç–∞ (–≥—Ä—É–ø–ø—ã)
async def read_chat(chat_id):
    async with TelegramClient('anon', telegram_key.API_ID, telegram_key.API_HASH) as client:
        
        # –ê–≤—Ç–æ—Ä–∏–∑—É–µ–º—Å—è
        await client.start(telegram_key.PHONE)
        
        #me = await client.get_me()
        #print(me.stringify())

        async for message in client.iter_messages(chat_id, limit=20):
            sender = await message.get_sender()
            if isinstance(message.sender,User):
                print(f'–û—Ç: {sender.first_name}')
            else:
                print(f'–û—Ç: {sender.title}')
            print(f'–¢–µ–∫—Å—Ç: {message.text}')
            print(f'–î–∞—Ç–∞: {message.date}')
            print('-' * 50)
```
